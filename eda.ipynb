Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv("../data/train.csv")

# Basic stats
print(train.info())
print(train.describe())

# 10 Visualizations
plt.figure(figsize=(6,4))
sns.countplot(x='is_claim', data=train)
plt.title("Target Distribution")
plt.show()

sns.histplot(train['age_of_policyholder'], bins=30, kde=True)
plt.title("Age of Policyholder Distribution")
plt.show()

sns.histplot(train['age_of_car'], bins=30, kde=True)
plt.title("Age of Car Distribution")
plt.show()

sns.boxplot(x='is_claim', y='policy_tenure', data=train)
plt.title("Policy Tenure vs Claim")
plt.show()

sns.countplot(x='fuel_type', hue='is_claim', data=train)
plt.title("Fuel Type vs Claim")
plt.show()

sns.countplot(x='segment', hue='is_claim', data=train)
plt.title("Car Segment vs Claim")
plt.show()

sns.countplot(x='transmission_type', hue='is_claim', data=train)
plt.title("Transmission vs Claim")
plt.show()

sns.countplot(x='ncap_rating', hue='is_claim', data=train)
plt.title("NCAP Rating vs Claim")
plt.show()

sns.scatterplot(x='population_density', y='age_of_policyholder', hue='is_claim', data=train)
plt.title("Population Density vs Age of Policyholder")
plt.show()

corr = train.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
